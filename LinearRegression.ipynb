```python
#importing necessary dependacies
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

```python
#importing housing data set from sklearn
Data = fetch_california_housing(as_frame=True)
```

```python
#transforming the dataset into a dataframe. We have actual data X, features and target variable Y.
#transforming the dataset to get what is necessary. Independent values X and dependent values Y.
df_x = pd.DataFrame(Data.data, columns=Data.feature_names)
df_y = pd.DataFrame(Data.target)
```

```python
#to get the statistics of the data that we have transformed
df_x.describe()
```

```python
df_y.describe()
```

```python
#initaialize the linear regression model
reg = LinearRegression()
```

```python
#splitting the data into 70% training and 30% testing data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=0)
```

```python
#training our model with training data
reg.fit(x_train,y_train)
```

```python
#printing the coeffs of the equation
print(reg.coef_)
```

```python
#print the predictions on test data
y_pred = reg.predict(x_test)
```

```python
y_pred
```

```python
#print the actual values
y_test
```

```python
#measuring the performance using Mean Squared Error using sklearn.metrics
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_test,y_test, "o", color ="green")
plt.plot(x_test,y_pred, color = "orange")
plt.title("Prediction")
plt.xlabel("House area")
plt.ylabel("Selling price")
```
