```python
#Gradient descent alogorithm for estimating m and c values
def Gradient_descent(m_now, c_now, points, L):
    m_new = 0
    c_new = 0
    n = len(points)
    for i in range (n):
        x_point = points.iloc[i].x
        y_point = points.iloc[i].y
        m_new += -(2/n) * x_point * (y_point-(m_now * x_point + c_now))
        c_new += -(2/n) * (y_point-(m_now * x_point + c_now))

    m_new = m_now - (L * m_new)
    c_new = c_now - (L * c_new)
    return m_new, c_new
```
